```python
import sympy as sp
from sympy.interactive import printing
printing.init_printing(use_latex=True)
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
t = sp.Symbol('t')
y = sp.Function('y')(t)

y0 = 2
y1 = 0
equation = sp.Eq(y.diff(t,2)+2*y.diff(t)+6*y,0)

solution = sp.dsolve(equation, ics = {y.subs(t,0):2})
print('(a)')
display(solution)

print('(b)')
solution = sp.dsolve(equation, ics = {y.subs(t,0):2,y.subs(t,1):0})
yp = solution.rhs.diff(t)
fyp = sp.lambdify(t,yp)
print("\t\t\talpha = {:.3f}".format(fyp(0)))

print('(c)')
a = sp.Symbol('alpha')
solution = sp.dsolve(equation, ics = {y.diff(t).subs(t,0):a, y.subs(t,0):2})
equation2 = sp.Eq(0,solution.rhs)
solvet = sp.solve(equation2, t)
#print("t : ")
#display(solvet)
fuctionofa = sp.lambdify(a,solvet)
T = sp.Function('T')(a)
tmin = sp.Eq(T,1/sp.sqrt(5)*(-sp.atan(2*sp.sqrt(5)/(a+2))+sp.pi))
print('tmin : ')
display(tmin)
#plt.plot(np_t,fuctionofa(np_t))

print('(d)')
lim = sp.limit(tmin.rhs, a, math.inf)
display(lim)

'''
t = sp.Symbol('t')
solution = sp.dsolve(equation, ics = {y.subs(t,0):2,y.subs(t,1):0})
np_t = np.linspace(-1,5,10000)
np_f = sp.lambdify(t,solution.rhs)
plt.plot(np_t,np_f(np_t))
'''

```
