```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
t = sp.symbols('t')
y = sp.Function('y')(t)

eqa = sp.Eq(y.diff(t).diff(t) - 6*y.diff(t) + 9*y, 0)
sol = sp.dsolve(eqa, y, ics = {y.subs(t,0):0, y.diff(t).subs(t,0):3})
sp.plot(sol.rhs, ylim=(0,10000))
f = sp.lambdify(t,sol.rhs)
x = np.linspace(-1,1)
plt.plot(x,f(x))
```
