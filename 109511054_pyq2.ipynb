```python
import scipy.integrate as sci
import math
from math import sin, cos
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display


def model(U,t):
  return [U[1],t*t + 2*cos(t)]

t = np.linspace(0,1)
U0 = [0,1]
ypp = sci.odeint(model,U0,t)
#print(ypp)
y = ypp[:,0]
ys = ypp[:,1]
plt.figure(1)
plt.xlabel('t')
plt.ylabel('y')
plt.plot(t,y)
plt.figure(2)
plt.xlabel('y')
plt.ylabel('dydt')
plt.plot(y,ys)
```

```python
import sympy as sp
from sympy import cos
import numpy as np
import matplotlib.pyplot as plt


t = sp.Symbol('t')
y = sp.Function('y')(t)

eqa = sp.Eq(y.diff(t,2), t*t + 2*cos(t))
sol = sp.dsolve(eqa, y, ics = {y.subs(t,0):0, y.diff(t).subs(t,0):1})
display(sol)


eqa2 = sp.Eq(y.diff(t), t*t + 2*cos(t))
sol2 = sp.dsolve(eqa2, y, ics = {y.subs(t,0):1})
eqadisplay = sp.Eq(y.diff(t),sol2.rhs)
display(eqadisplay)


display(eqa)

lam1 = sp.lambdify(t, sol.rhs)
lam2 = sp.lambdify(t, sol2.rhs)
t = np.linspace(0,1)

plt.figure(1)
plt.plot(t,lam1(t))
plt.figure(2)
plt.plot(lam1(t),lam2(t))
```

```python
import scipy.integrate as sci
from sympy import sin
import math
import numpy as np
import matplotlib.pyplot as plt


def model(y,t):
  return -2*sin(t)-t/10*(math.e**(-t/10))

lt = np.linspace(-10,10)

y01 = -2
ly1 = sci.odeint(model, y01, lt)
y0 = -1
ly2 = sci.odeint(model, y0, lt)
y0 = 0
ly3 = sci.odeint(model, y0, lt)
y0 = 1
ly4 = sci.odeint(model, y0, lt)
y0 = 2
ly5 = sci.odeint(model, y0, lt)


plt.xlim(-10,10)
plt.ylim(-10,20)
plt.plot(lt,ly1, label = 'y(0) = -2')
plt.plot(lt,ly2, label = 'y(0) = -1')
plt.plot(lt,ly3, label = 'y(0) = 0')
plt.plot(lt,ly4, label = 'y(0) = 1')
plt.plot(lt,ly5, label = 'y(0) = 2')
plt.legend()

lt = np.linspace(-10,10,20)
ly = np.linspace(-10,20,20)
ltt, lyy = np.meshgrid(lt,ly)
lm = -2*np.sin(ltt)-ltt/10*(math.e**(-ltt/10))
xx = np.cos(np.arctan(lm))
yy = np.sin(np.arctan(lm))
plt.quiver(ltt,lyy,xx,yy)
```
